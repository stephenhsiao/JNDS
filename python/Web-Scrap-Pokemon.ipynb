### Use beautifulsoup to pull the pokemon's name, number, and fighting type out of the elements in the table

```python
import urllib2
from bs4 import BeautifulSoup
import json
import pandas as pd

soup = BeautifulSoup(urllib2.urlopen('http://www.serebii.net/diamondpearl/pokemon.shtml').read())
```

### Put the pokemon data into a dataframe and print it

```python
tables = soup.find_all('table', width = "98%")
```

```python
raw = {
    'number':[],
    'names':[],
    'type1':[],
    'type2':[]
}

for idx, table in enumerate(tables):
    for idx2, row in enumerate(table.find_all('tr')):
        if idx2 > 0:
            for idx3, column in enumerate(row.find_all('td')):
                if idx3 == 0:
                    raw['number'] += [column.text.replace('\r', '').replace('\n', '').replace('\t','').replace('#','')]
                if idx3 == len(row.find_all('td')) - 4:
                    raw['names'] += [column.text.replace('\n','')]
                if idx3 == len(row.find_all('td'))-2:
                    string = column.find('a').contents[0]['src']
                    raw['type1'] +=  [string.replace('/pokedex-dp/type/', "").replace('.gif', '')]
                if idx3 == len(row.find_all('td'))-1: 
                    string = column.find('a').contents[0]['src']
                    raw['type2'] +=  [string.replace('/pokedex-dp/type/', "").replace('.gif', '')]
```

```python
# Put the dictionary's data into a DataFrame
ndf = pd.DataFrame(raw)
ndf['fightingType'] = ndf['type1'] + '-' + ndf['type2']

for (i,x) in enumerate(ndf['fightingType']):
    ndf['fightingType'][i] = x.replace('-na', '')
    
ndf.drop(['type1', 'type2'], axis = 1, inplace =True)
ndf
```
