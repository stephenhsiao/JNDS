# Support Vector Machines
- To solve binary classification problem
- Define the concept of an optimal linear separating hyperplane - a linear classifier known as a maximal margin classifier (MMC). 
- We will then relax the restriction of linearity and consider non-linear classifiers, namely support vector machines （SVM）, which use kernel functions to improve computational efficiency

````python
plt.figure(figsize=(10,6))
ax = plt.gca()
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('X1', fontsize=14)
ax.set_ylabel('X2', fontsize=14)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap=plt.cm.flag, edgecolors='black', linewidth=1);

xfit = np.linspace(-1, 3.5)

for m, b in [(1, 0.65), (0.5, 1.6), (-0.2, 2.9), (.3, 2.3), (.7, 1)]:
    plt.plot(xfit, m * xfit + b, '-k')
````

## But what if we took each of those lines and made them really fat?

## How fat?

## Up until they were jammed in between the points on either side like this...

````python
plt.figure(figsize=(10,6))
ax = plt.gca()
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('X1', fontsize=14)
ax.set_ylabel('X2', fontsize=14)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap=plt.cm.flag, edgecolors='black', linewidth=1);

xfit = np.linspace(-1, 3.5)

for m, b, d in [(1, 0.60, 0.4), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.24)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none', color='lightgrey', alpha=0.4)

plt.xlim(-1, 3.5);
````

## If we did that and tried a bunch of lines, we could compare the lines to find the fattest line we could draw between each class.

## That is exactly what a support vector machine is.

````python
plt.figure(figsize=(10,6))
ax = plt.gca()
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('X1', fontsize=14)
ax.set_ylabel('X2', fontsize=14)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap=plt.cm.flag, edgecolors='black', linewidth=1);

xfit = np.linspace(-1, 3.5)

for m, b, d in [(.1, 2.4, 0.65)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none', color='lightgrey', alpha=0.4)

plt.xlim(-1, 3.5);
````

We can now see that the center line is the "thickest" of the three. If this was
the thickest of all possible lines, we would have maximized the margins between
the decision boundary between and the two classes. The margins being the points
nearest to the decision or separating boundary. Those points on the edge are called the "support vectors" as they appear to support that plane. 

In the terminology of SVMs, we have created a **maximum margin hyperplane**
(an n-dimensional plane).

## Check: How sensitive to outliers would a support vector machine be?

## Let's use sklearn to fit one

````python
from sklearn.svm import SVC
clf = SVC(kernel='linear')
clf.fit(X, y)
````

## Now we can visualize it

````python
def plot_svc_decision_function(clf, ax=None):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    x = np.linspace(plt.xlim()[0], plt.xlim()[1], 30)
    y = np.linspace(plt.ylim()[0], plt.ylim()[1], 30)
    Y, X = np.meshgrid(y, x)
    P = np.zeros_like(X)
    for i, xi in enumerate(x):
        for j, yj in enumerate(y):
            P[i, j] = clf.decision_function(np.array([xi, yj]).reshape(1, -1))
    # plot the margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
plt.figure(figsize=(10,6))
ax = plt.gca()
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('X1', fontsize=14)
ax.set_ylabel('X2', fontsize=14)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap=plt.cm.flag, edgecolors='black', linewidth=1)

plot_svc_decision_function(clf);
````

## And, here, you can see the support vectors

````python
plt.figure(figsize=(10,6))
ax = plt.gca()
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('X1', fontsize=14)
ax.set_ylabel('X2', fontsize=14)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap=plt.cm.flag, edgecolors='black', linewidth=1)
plot_svc_decision_function(clf)
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=300, facecolors='none', cmap=plt.cm.flag, edgecolors='black', linewidth=1);
````

## What do we do if the data is not linearly separable?

### Cut us some slack (variables)...

<br>If we cannot linearly separate our data, we can still attempt to maximize the margin but subject to a a penalty for each misclassified example. This says we are willing to make a 
tradeoff between the size of the margin and the misclassified samples. This is called a
**soft margin** classifier as opposed to a **hard margin** one.

We can adjust this penalty through the C parameter in sklearn. The default value is 1.

## Let's see how changing this penalty affects the model

````python
from sklearn.svm import SVC
clf = SVC(kernel='linear', C=1)
clf.fit(X, y)
````

````python
from sklearn.preprocessing import Imputer, StandardScaler

def plot_svc_decision_function(clf, ax=None):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    x = np.linspace(plt.xlim()[0], plt.xlim()[1], 30)
    y = np.linspace(plt.ylim()[0], plt.ylim()[1], 30)
    Y, X = np.meshgrid(y, x)
    ss = StandardScaler()
    X = ss.fit_transform(X)
    P = np.zeros_like(X)
    for i, xi in enumerate(x):
        for j, yj in enumerate(y):
            P[i, j] = clf.decision_function(np.array([xi, yj]).reshape(1, -1))
    # plot the margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
plt.figure(figsize=(10,6))
ax = plt.gca()
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('X1', fontsize=14)
ax.set_ylabel('X2', fontsize=14)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap=plt.cm.flag, edgecolors='black', linewidth=1)

plot_svc_decision_function(clf);
````

You can see we have the one point misclassified and the margin is roughly the same.

## So, what is a kernel?

## Comparison between logistic regression and SVM:

- The basic idea of logistic regression is to adapt linear regression so that it estimates the probability a new entry falls in a class.
- In short, they have different objectives. Logistic regression finds a classifier which maximizes the conditional likelihood of the training data. SVM maximizes the margin between points closest to the classification boundary.
- SVM's are actually formulated so that ONLY points near the decision boundary really make a difference. Points that are "obvious" have no effect on the decision boundary. This can be very different from logistic regression, as "obvious" points can sometimes be very influential, since every training point has a certain influence on the estimated logistic regression function.
